```python
import pandas as pd
import requests
```

```python
APPL_JSON_1 = requests.get('https://api.polygon.io/v2/aggs/ticker/AAPL/range/5/minute/2022-07-18/2022-07-22?adjusted=true&sort=asc&limit=5000&apiKey=7FNn1a46_CksQGa1IfQqAeloyYEjruY3')
```

```python
import json
import pathlib
pathlib.Path('data.json').write_bytes(APPL_JSON_1.content)
df = pd.read_json('data.json')
df['results'][0]['c']
c, h ,l,n,o,t,v,vw = [],[],[],[],[],[],[],[]
i = 0
for j in df.iterrows():
    c.append(df['results'][i]['c'])
    h.append(df['results'][i]['h'])
    l.append(df['results'][i]['l'])
    n.append(df['results'][i]['n'])
    o.append(df['results'][i]['o'])
    t.append(df['results'][i]['t'])
    v.append(df['results'][i]['v'])
    vw.append(df['results'][i]['vw'])
    i+= 1
```

```python
df['Open'] = o
df['Close'] = c
df['Low'] = l
df['High'] = h
df['Number of Transactions'] = n
df['Time'] = t
df['Volume'] = v
df['Vol Weighted Average Price'] = vw
df
```

```python
df = df.drop(['results'],axis = 1)
df
```
